```python
#importing all libraries
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import warnings
warnings.filterwarnings('ignore')
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
```

```python
#importing the data via pandas read csv
df = pd.read_csv('vaccination_tweets.csv')
df.head() #checking the imported data
```

```python
df.info() #data info
```

```python
df.isnull().sum() #total number of missing columns or null values in the data
```

```python
df.columns #all the columns present in the data
```

```python
#dropping all columns except text column, since we only need it to perform the sentiment analysis
text_df = df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'],axis=1)
text_df.head()
```

```python
text_df.info() #new text data info
```

```python
#preprocessing the text data
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '', text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing) #applying the preprocessing function on the text data
```

```python
text_df = text_df.drop_duplicates('text') #eliminating all duplicated rows in the text data
```

```python
#stemming operation
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
text_df.info()
```

```python
#determining polarity on each of the sentences in the dataset
#polar words and phrases are usually associated with strong or clearly defined sentiments while non-polar words are refers to words and phrases
#that are typically used in everyday mundane language
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head(10)
```

```python
#The target variable, the sentiment.
def sentiment(label):
    if label < 0:
        return "Negative"
    elif label == 0:
        return "Neutral"
    elif label > 0:
        return "Positive"
```

```python
#determining the sentiment categorical values based on polarity values
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
text_df.head()
```

```python
#visualizing the sentiment values with countplot bar chart
fig = plt.figure(figsize=(5,5))
sns.countplot (x = 'sentiment', data=text_df)
```

```python
#data processing with count vectorizer
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of feature: {}\n".format(len(feature_names)))
print("First 20 feature:\n{}".format(feature_names[:20]))
```

```python
X = text_df['text']
Y = text_df['sentiment']
x = vect.transform(X)
```

```python
#splitting the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, Y, test_size=0.2, random_state=42)
```

```python
#Building the model
#Logistic Regression classification model
lg_model = LogisticRegression()
lg_model.fit(x_train, y_train)
prediction = lg_model.predict(x_test)
accuracy = accuracy_score(prediction, y_test)
print("Test accuracy: {:.2f}%".format(accuracy*100))
```

```python
#Visualizing logistic regression performance via confusion matrix
xticks=['Negative', 'Neutral', 'Positive']
yticks=['Negative', 'Neutral', 'Positive']
conf_m = confusion_matrix(y_test, prediction)
print('Confusion Matrix:\n',conf_m)
plt.figure()
plt.title('Confusion Matrix for Logistic Regression', fontsize=10)
sns.set(font_scale=1)
ax=sns.heatmap(conf_m, fmt='d', xticklabels=xticks, yticklabels=yticks, annot=True, cmap='Reds')
ax.set_ylim([0,3])
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```

```python
#Linear SVC model
svc_model = LinearSVC()
svc_model.fit(x_train, y_train)
svc_prediction = svc_model.predict(x_test)
svc_accuracy = accuracy_score(svc_prediction, y_test)
print("Test accuracy: {:.2f}%".format(svc_accuracy*100))
```

Linear SVC model seemed to have performed better than the Logistic Regression model

```python
#visualizing performance via confusion matrix
xticks=['Negative', 'Neutral', 'Positive']
yticks=['Negative', 'Neutral', 'Positive']
svc_conf_m = confusion_matrix(y_test, svc_prediction)
print('Confusion Matrix:\n',svc_conf_m)
plt.figure()
plt.title('Confusion Matrix for Linear SVC model', fontsize=10)
sns.set(font_scale=1)
ax=sns.heatmap(svc_conf_m, fmt='d', xticklabels=xticks, yticklabels=yticks, annot=True, cmap='Blues')
ax.set_ylim([0,3])
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```

```python
#Hyperparameter tuning with GridSearchCV
#optimizing the Linear SVC model
param_grid = {'C':[0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(svc_model, param_grid)
grid.fit(x_train, y_train)
```

```python
print("Best parameter:", grid.best_params_)
```

```python
grid_search_prediction = grid.predict(x_test)
grid_search_accuracy = accuracy_score(grid_search_prediction, y_test)
print("Test accuracy: {:.2f}%".format(grid_search_accuracy*100))
```

Optimized model performed better than the original by a few decimal percentage
